```python
# To change Age_code  months into years

# Copy dataframe data to df1 where age _code in months
df1 = df[df['Age_Code'] =='Months']
print(df1)
```

```python
df1[['Age','Age_Code']]
```

```python
#Convert Age months to years
df1['Age']=df1['Age']/12
```

```python
df1['Age_Code']='Year'
```

```python
#checking dataframe after conversion 
df1[['Age','Age_Code']]
```

```python
#Create a new dataframe df2 here age_code is year 
df2=df[df['Age_Code'] == 'Year']
print(df2)
```

```python
#concatenate df2 and df1 in new dataframe df3
df3 = pd.concat([df2,df1])
```
