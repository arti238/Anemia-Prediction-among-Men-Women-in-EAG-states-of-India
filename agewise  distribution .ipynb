```python
data = df[['Age','Age_Code']]
print(data)
```

```python
#dropping days wise age columns 
data = data.drop(data.loc[data['Age_Code']=='Days'].index)
print(data)
```

```python
# To change Age months into years

# Copy dataframe data to df2 where age _code in months
df2 = data[data['Age_Code'] =='Months']
```

```python
df2[['Age','Age_Code']]
```

```python
#Convert Age months to years
df2['Age']=df2['Age']/12
```

```python
df2['Age_Code']='Year'
```

```python
#checking dataframe after conversion 
df2[['Age','Age_Code']]
```

```python
#Create a new dataframe df3 where age_code is year 
df3=data[data['Age_Code'] =='Year']
print(df3)
```

```python
#concatenate df2 and df3 in new dataframe df4
df4 = pd.concat([df3,df2])
```

```python
df4
```

```python
# selecting rows based on condition(age range between 0-3years) 
df5 = df4[(df4['Age'] >=0) &(df4['Age']<=3)]
print(df5)
```
